# Data Science Job Salaries
### Analysis of data on  Data science job salaries downloaded from kaggle "https://www.kaggle.com/datasets/ruchi798/data-science-job-salaries" gathered by Ruchi Bhatia.
### Python on jupyter notebook is used to do the following:
### Read the downloaded data from kaggle
### Add new columns with easily understod descriptions for the abbreviation used in the data.
### Write the modified data to a CSV file. (That is then imported to a local SQL server using Microsoft SQL Server Management Studio 18)
### Access data from the local SQL server and save the data to a CSV file. (That is then imported to MongoDB using MongoDB compass)
### 
### MongoDB is used to analyse and visualise hte data.

### Link to MongoDB to the visualisation of the data:

```python
# Need pandas to read/write csv files and modify them
import pandas as pd

# Need pyodbc to interact with SQL Server
import pyodbc
```

###### work_year = The year the salary was paid.
###### experience_level = The experience level in the job during the year with the following possible values: EN Entry-level / Junior MI Mid-level / Intermediate SE Senior-level / Expert EX Executive-level / Director
###### employment_type = The type of employement for the role: PT Part-time FT Full-time CT Contract FL Freelance
###### job_title = The role worked in during the year.
###### salary = The total gross salary amount paid.
###### salary_currency = The currency of the salary paid as an ISO 4217 currency code.
###### salaryinusd = The salary in USD (FX rate divided by avg. USD rate for the respective year via fxdata.foorilla.com).
###### employee_residence =  Employee's primary country of residence in during the work year as an ISO 3166 country code.
###### remote_ratio = The overall amount of work done remotely, possible values are as follows: 0 No remote work (less than 20%) 50 Partially remote 100 Fully remote (more than 80%)
###### company_location = The country of the employer's main office or contracting branch as an ISO 3166 country code.
###### company_size = The average number of people that worked for the company during the year: S less than 50 employees (small) M 50 to 250 employees (medium) L more than 250 employees (large)


```python
# Reading ds_salaries.csv data that is downloaded from kaggle "https://www.kaggle.com/datasets/ruchi798/data-science-job-salaries" gathered by Ruchi Bhatia
df_ds_salaries = pd.read_csv('datafolder/ds_salaries.csv', encoding='latin-1', index_col=0)
df_ds_salaries
```

```python
# Creating new column with name "exp_level" and giving it default input "Unemp". 
# Then for looping with range and giving column "exp_level" a clear text descrition of the	abbreviation in column "experience_level"
df_ds_salaries['exp_level'] = 'Unemp'
for x in range(607):
  if df_ds_salaries['experience_level'].iloc[x] == 'EN':
    df_ds_salaries['exp_level'].iloc[x] = 'Entry-level'
  elif df_ds_salaries['experience_level'].iloc[x] == 'MI':
    df_ds_salaries['exp_level'].iloc[x] = 'Mid-level'
  elif df_ds_salaries['experience_level'].iloc[x] == 'SE':
    df_ds_salaries['exp_level'].iloc[x] = 'Senior-level'
  elif df_ds_salaries['experience_level'].iloc[x] == 'EX':
    df_ds_salaries['exp_level'].iloc[x] = 'Executive-level'
  elif df_ds_salaries['experience_level'].iloc[x] == 'Director':
    df_ds_salaries['exp_level'].iloc[x] = 'Director'
```

```python
# Creating new column with name "exp_type" and giving it default input "Emp". 
# Then for looping with range and giving column "emp_type" a clear text descrition of the	abbreviation in column "employment_type"
df_ds_salaries['emp_type'] = 'Emp'
for x in range(607):
  if df_ds_salaries['employment_type'].iloc[x] == 'PT':
    df_ds_salaries['emp_type'].iloc[x] = 'Part-time'
  elif df_ds_salaries['employment_type'].iloc[x] == 'FT':
    df_ds_salaries['emp_type'].iloc[x] = 'Full-time'
  elif df_ds_salaries['employment_type'].iloc[x] == 'CT':
    df_ds_salaries['emp_type'].iloc[x] = 'Contract'
  elif df_ds_salaries['employment_type'].iloc[x] == 'FL':
    df_ds_salaries['emp_type'].iloc[x] = 'Freelance'
```

```python
# Creating new column with name "salary_in_eur" and giving it default input "0". Defining exchange rate from USD to EUR as usd_to_eur.
# Then for looping with range calculates the salary in EUR, from column "salary_in_usd" with the exchange rate, and insert it in column "salary_in_eur"
df_ds_salaries['salary_in_eur'] = 0
usd_to_eur = float(0.97)
for x in range(607):
  salaries_euro_float = ((df_ds_salaries['salary_in_usd'].iloc[x])*usd_to_eur)
  df_ds_salaries['salary_in_eur'].iloc[x] = int(salaries_euro_float)
```

```python
# Checking that the modified table looks correct
df_ds_salaries
```

```python
# Writing the modified table "df_ds_salaries" to data folder as ds_salaries_added_edit.csv
df_ds_salaries.to_csv('datafolder/ds_salaries_added_edit.csv',index=False)
```

### I used Microsoft SQL Server Management Studio 18 to import the file "ds_salaries_added_edit.csv" to my local SQL server

```python
# Defined connectionstring to connect to the local SQL servers database with the data science job salaries data
connectionstring = 'DRIVER={ODBC Driver 17 for SQL Server};Server=localhost\\SQLEXPRESS;Database=DataScienceJobSalaries;trusted_connection=yes'

# Creating context between python and the SQL database
context = pyodbc.connect(connectionstring)

# Creating cursor to use the SQL database through context
cursor = context.cursor()

# Defining a SQL command to get data for all columns where column "job_tilte" has "Data Engineer" as its value
# The data handeled is quite small and I do not find it nesessary to limit accessed columns.
sqlcommand = "SELECT * FROM dbo.ds_salaries_added_edit WHERE job_title='Data Engineer';"

# Executing the SQL command to retrive data from the SQL server
cursor.execute(sqlcommand)

mydataset = []

results = cursor.fetchall()
for result in results:
    dummy = list(result)
    mydataset.append(dummy)

# Creating dataframe with the name "df_job_title_data_engineer" of the retrived data from the SQL server
mycolumn = ['work_year','experience_level','employment_type','job_title','salary','salary_currency','salary_in_usd','employee_residence','remote_ratio','company_location','company_size','exp_level','emp_type','salary_in_eur']
df_job_title_data_engineer = pd.DataFrame(mydataset,columns=mycolumn)

# Cheking the retrived data
df_job_title_data_engineer
```

```python
# Writing the modified table "df_job_title_data_engineer" to data folder as ds_salaries_data_engineer.csv
df_job_title_data_engineer.to_csv('datafolder/ds_salaries_data_engineer.csv',index=False)
```
